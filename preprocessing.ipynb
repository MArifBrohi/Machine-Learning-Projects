```python
import pandas as pd
import re
from datetime import datetime
```

```python
f = open('/content/drive/MyDrive/Machine Learning Project/WhatsApp Chat with WEB engineering section B.txt', encoding='utf-8')
```

```python
data = f.read()
print(data)
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{1,2},\s\d{1,2}:\d{1,2}\s\w+\s-\s'
```

```python

dates = re.findall(pattern,data)
```

```python
dates
```

```python

modified_dates = [s.replace('\u202f','').replace(',','').replace(' - ','') for s in dates]

# modified_dates = [s.replace(',',' ') for s in modified_dates]

modified_dates
# print(len(modified_dates))
# print(type(modified_dates))
```

```python
messages = re.split(pattern,data)[1:]
messages
# print(len(messages))
```

```python
df = pd.DataFrame({'message':messages, 'date':modified_dates})

df['date']  = pd.to_datetime(df['date'],format='%m/%d/%y %I:%M%p')
df['date'] = df['date'].dt.strftime('%Y-%m-%d %I:%M:%S %p')
# df.rename(columns={'msg_dates':'date'}, inplace=True)
df
```

```python
user_column = []
message_column = []
for msg in df['message']:
  entry = re.split('([\w\W]+?):\s', msg)
  if entry[1:]: # checks whether our messages are splited successfully
    user_column.append(entry[1])
    message_column.append(entry[2])
  else:
    user_column.append('group_notification')
    message_column.append(entry[0])

df['User'] = user_column
df['Message'] = message_column
df.drop(columns=['message'],inplace=True)
df.head(20)
```

```python
date = []
time = []

for date_time in df['date']:
    # Use a regular expression to match the date and time components
    entry = re.split(r'(\d{4}-\d{2}-\d{2}) (\d{2}:\d{2}:\d{2} [APMapm]{2})', date_time)

    if entry[1:]:
        date.append(entry[1])
        time.append(entry[2])
    else:
        date.append('')
        time.append('')

df['Date'] = date
df['Time'] = time
df.drop(columns=['date'], inplace=True)
df.head()
```

#Monthly Timeline

```python
dates = df['Date']
dates.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

# Extract year, month, and day
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day

# Create a new column for month names
df['Month_Name'] = df['Date'].dt.strftime('%B')

# Display the DataFrame
print(df)
```

```python
dates = df['Date']

# Assuming dates is your list of strings

# Extract years, months, and days using a list comprehension
years = [date[:4] for date in dates]
months = [date[5:7] for date in dates]
days = [date[8:10] for date in dates]

print("Extracted Years:")
print(years)

print("Extracted Months:")
print(months)

print("Extracted Days:")
print(days)

```

```python

df['Year'] = years
df['Month'] = months
df['Day'] = days
```

```python
df
```

```python
timeline = df.groupby(['Year','Month','Month_Name','Day']).count()['Message'].reset_index()
```

```python
timeline
```

```python
time = []
for i in range(timeline.shape[0]):

  time.append(timeline['Month_Name'][i]+'-'+str(timeline['Year'][i]))
```

```python
timeline['Time'] = time
```

```python
timeline
```

```python
plt.plot(timeline['Time'],timeline['Message'])
plt.xticks(rotation='vertical')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(timeline['Time'], timeline['Message'], color='skyblue')
plt.xticks(rotation='vertical')
ax.set_ylabel('Message')
ax.set_xlabel('Time')
plt.show()
```
